```python
labels = ['view', 'cart','purchase']
size = data['event_type'].value_counts()
colors = ['red', 'yellow','blue']
explode = [0, 0.1,0.1]

plt.rcParams['figure.figsize'] = (8, 8)
plt.pie(size, colors = colors, explode = explode, labels = labels, shadow = True, autopct = '%.2f%%')
plt.axis('off')
plt.legend()
plt.show()
```
